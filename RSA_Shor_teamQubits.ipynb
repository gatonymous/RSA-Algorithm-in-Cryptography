# Generación de numeros primos

````python
import random

def generar_numero(bits):
    # Genera un número aleatorio con el número de bits especificado
    return random.getrandbits(bits) | (1 << (bits - 1)) | 1

def es_primo(n, k=40):
    # Implementa el test de primalidad de Miller-Rabin
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False

    # Escribe n-1 como d*2^r
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    # Realiza k pruebas de Miller-Rabin
    for _ in range(k):
        a = random.randint(2, n - 2)
        x = pow(a, d, n)
        if x == 1 or x == n - 1:
            continue
        for _ in range(r - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False
    return True

def generar_primo(bits):
    while True:
        candidato = generar_numero(bits)
        if es_primo(candidato):
            return candidato
````

````python
# Ejemplo de uso: generar un primo de 512 bits
bits = 512
primo = generar_primo(bits)
print(f"Primo de {bits} bits: {primo}")
````

````python
# Función para calcular el máximo común divisor
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Función para encontrar el inverso modular
def modinv(a, m):
    m0, x0, x1 = m, 0, 1
    if m == 1:
        return 0
    while a > 1:
        q = a // m
        m, a = a % m, m
        x0, x1 = x1 - q * x0, x0
    if x1 < 0:
        x1 += m0
    return x1
````

````python
# Calcular n y phi
n = p * q
phi = (p - 1) * (q - 1)

# Elegir un número e tal que 1 < e < phi y gcd(e, phi) = 1
e = random.randrange(1, phi)
g = gcd(e, phi)
while g != 1:
    e = random.randrange(1, phi)
    g = gcd(e, phi)

# Calcular d, el inverso modular de e
d = modinv(e, phi)

# Claves pública y privada
public_key = (e, n)
private_key = (d, n)

print("Clave pública: ", public_key)
print("Clave privada: ", private_key)

````

````python
# Función para encriptar un mensaje
def encrypt(plaintext, key):
    e, n = key
    ciphertext = [pow(ord(char), e, n) for char in plaintext]
    return ciphertext

# Función para desencriptar un mensaje
def decrypt(ciphertext, key):
    d, n = key
    plaintext = ''.join([chr(pow(char, d, n)) for char in ciphertext])
    return plaintext

````

````python
# Mensaje a encriptar
message = "Hola, mundo!"

# Encriptar el mensaje
encrypted_msg = encrypt(message, public_key)
print("Mensaje encriptado: ", encrypted_msg)

# Desencriptar el mensaje
decrypted_msg = decrypt(encrypted_msg, private_key)
print("Mensaje desencriptado: ", decrypted_msg)
````

## Algoritmos optimizados

````python
def gcd_stein(a, b):
    if a == b:
        return a
    if a == 0:
        return b
    if b == 0:
        return a

    # Reduce even numbers
    if ~a & 1:  # a is even
        if b & 1:  # b is odd
            return gcd_stein(a >> 1, b)
        else:  # both are even
            return gcd_stein(a >> 1, b >> 1) << 1
    if ~b & 1:  # a is odd, b is even
        return gcd_stein(a, b >> 1)

    # Reduce larger number
    if a > b:
        return gcd_stein((a - b) >> 1, b)
    return gcd_stein((b - a) >> 1, a)
````

## Algoritmo paralelizado (Periodo de Shor)

````python
import multiprocessing

# Obtener el número de CPUs disponibles
num_cpus = multiprocessing.cpu_count()

print(f"Número de hilos (threads) disponibles: {num_cpus}")
````

````python
import concurrent.futures
import math
import time

# Función para encontrar el período en un subrango específico
def find_period_in_range(a, N, start, end):
    x = 1
    for k in range(start, end):
        x = (x * a) % N
        if x == 1:
            return k
    return None

# Función principal para encontrar el período utilizando paralelización
def find_period_parallel(a, N, num_threads=4):
    # Determinar el rango de valores para cada hilo
    step = N // num_threads
    ranges = [(i * step + 1, (i + 1) * step + 1) for i in range(num_threads)]

    # Ejecutar la búsqueda en paralelo
    with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
        futures = [executor.submit(find_period_in_range, a, N, start, end) for start, end in ranges]
        for future in concurrent.futures.as_completed(futures):
            result = future.result()
            if result is not None:
                return result

    return None


````

# Decrypt with short classical algorithm

````python
import math
import random

# Función para encontrar el período r de la función a^x % N usando una búsqueda clásica
def find_period(a, N):
    x = 1
    k = 1
    while k < N:
        x = (x * a) % N
        if x == 1:
            return k
        k += 1
    return None

# Algoritmo clásico de Shor para la factorización
def shor_classical(N):
    # Paso 1: Seleccionar un número aleatorio a tal que 1 < a < N
    a = random.randint(2, N - 1)
    print(f"Elegido a: {a}")

    # Paso 2: Calcular el gcd de a y N
    d = gcd(a, N)
    if d > 1:
        return d, N // d

    # Paso 3: Encontrar el período r de a^x % N
    #r = find_period(a, N)
    r = find_period_parallel(a, N)
    if r is None or r % 2 != 0:
        return shor_classical(N)

    # Paso 4: Calcular los factores
    x = pow(a, r // 2, N)
    if x == N - 1:
        return shor_classical(N)

    p = gcd(x - 1, N)
    q = gcd(x + 1, N)

    if p == 1 or q == 1:
        return shor_classical(N)

    return p, q


````

````python
# Número a factorizar
# N = P*Q

# Ejecutar el algoritmo de Shor clásico
factors = shor_classical(N)
print(f"Factores de {N}: {factors}")
````
